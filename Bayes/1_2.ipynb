```python
from IPython.core.pylabtools import figsize
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
figsize(12.5, 4)
```

## 離散分布の場合
### Poission Disribution

```python

import scipy.stats as stats
poi = stats.poisson
lambda_ = [1.5, 4.25,12.2]
colors = ["#3423aD","#A93423","#C231AB"]
a = np.arange(16)
plt.bar(a, poi.pmf(a, lambda_[0]), color=colors[0], label="$\lambda$ = %0.1f" % lambda_[0], alpha=0.6, edgecolor=colors[0], lw=3)
plt.bar(a, poi.pmf(a ,lambda_[1]), color=colors[1], label="$\lambda$ = %0.1f" % lambda_[1], alpha=0.6, edgecolor=colors[1], lw=3)
plt.bar(a, poi.pmf(a, lambda_[2]), color=colors[2], label="$\lambda$ = %0.1f" % lambda_[2], alpha=0.6, edgecolor=colors[2], lw=3)
plt.xticks(a+0.4,a)
plt.xlabel("$k$")
plt.ylabel("Probability of $k$")
plt.title("Probability mass function of a Poisson random variables differing $\Lambda$ values")
plt.legend()
```

## 連続分布の場合
### Exponential Distribution

```python
a = np.linspace(0, 4, 100)
expo = stats.expon
lambda_ = [0.5,1,2]
```

```python
for L, C in zip(lambda_, colors):
    plt.plot(a, expo.pdf(a,scale=1/L),color=C, label="$\lambda=%.1f$" % L)
plt.xlabel("$z$")   
plt.ylabel("Probability density function at $z$")
plt.title("Exponential Distribution : $\lambda  \exp^{-\lambda z} (z > 0) $")
plt.legend()
```
